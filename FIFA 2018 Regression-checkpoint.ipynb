### Data visualization - Bar plots.

```python
Goal_score = []
for i in fifa_teams:
    x = fifa_df[fifa_df.Team == i]
    goal_score = sum(x['Goal Scored'])
    Goal_score.append(goal_score)

fifa_dict = pd.DataFrame({'Team': fifa_teams, 'Goal_score': Goal_score})
new_index = fifa_dict.Goal_score.sort_values(ascending=False).index.values
sorted_data = fifa_dict.reindex(new_index)

plt.figure(figsize = (30,10))
sns.barplot(x=sorted_data.Team,y=sorted_data.Goal_score)
plt.xlabel('Teams')
plt.ylabel('Total Goals in 2018')
plt.title('Summary of goals scored by various teams in 2018')
```

### Summary of goals scored by various teams in 2018.
##### 1. Belgium scored maximum goals 16 followed by Croatia and France.
##### 2. Iran, Morocco, Panama, Egypt and Saudi Arabia have scored lease goals in 2018.

```python
win_per = []
for i in fifa_teams:
    x = fifa_df[fifa_df.Team == i]
    win = len(x[x['Man of the Match'] == 'Yes']['Team'])
    win_perc = (win/len(x))*100
    win_per.append(win_perc)
   
fifa_dict = pd.DataFrame({'Team': fifa_teams, 'win_per': win_per})
new_index = fifa_dict.win_per.sort_values(ascending=False).index.values
sorted_data = fifa_dict.reindex(new_index)

plt.figure(figsize = (35,10))
sns.barplot(x=sorted_data.Team,y=sorted_data.win_per)
plt.xlabel('Teams')
plt.ylabel('Win % in 2018')
plt.title('Win percentage of various teams in FIFA 2018')
```

### Win percentage of various teams in FIFA 2018.
##### 1. France has won all the matches it played in 2018. Belgium, Brazil and Uruguay come next with win percentage between 75 - 80%.
##### 2. Japan, Mexico and Argentina have a very low win percentage.
##### 3. Panama, Australia, Costa Rica, Morocco and Saudi Arabia did not register a single win in 2018.

```python
Ball_poss_per = []
for i in fifa_teams:
    x = fifa_df[fifa_df.Team == i]
    ball_poss = sum(x['Ball Possession %']/len(x))
    Ball_poss_per.append(ball_poss)
   
fifa_dict = pd.DataFrame({'Team': fifa_teams, 'Ball_poss_per':Ball_poss_per})
new_index = fifa_dict.Ball_poss_per.sort_values(ascending=False).index.values
sorted_data = fifa_dict.reindex(new_index)

plt.figure(figsize = (35,10))
sns.barplot(x=sorted_data.Team,y=sorted_data.Ball_poss_per)
plt.xlabel('Teams')
plt.ylabel('Ball possession % of various teams')
plt.title('Average ball possession % of various teams in FIFA 2018')
```

### Ball possession %  of various teams in FIFA matches 2018.
##### 1. Spain, Germany and Argentina have the highest ball possession % on an average.
##### 2. Russia, Iceland and Iran have the lease ball possession %.

```python
pass_accuracy = []
for i in fifa_teams:
    x = fifa_df[fifa_df.Team == i]
    pass_acc = sum(x['Pass Accuracy %']/len(x))
    pass_accuracy.append(pass_acc)
   
fifa_dict = pd.DataFrame({'Team': fifa_teams, 'pass_accuracy':pass_accuracy})
new_index = fifa_dict.pass_accuracy.sort_values(ascending=False).index.values
sorted_data = fifa_dict.reindex(new_index)

plt.figure(figsize = (35,10))
sns.barplot(x=sorted_data.Team,y=sorted_data.pass_accuracy)
plt.xlabel('Teams')
plt.ylabel('Pass accuracy of various teams')
plt.title('Average pass accuracy of various teams in FIFA 2018')
```

### Pass accuracy % of various teams in FIFA 2018.
##### All the teams have pass accuracy of above 80% in FIFA 2018.

### Preparing the data to apply algoruthm 
##### 1. Preparing the training and test data sets
##### 2. Scaling the data

```python
X = fifa_df.loc[:,fifa_df.columns != 'Man of the Match']
X.head()
```

```python
Y = fifa_df['Man of the Match']
Y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.20, random_state=1)
```

```python
X_train.head()
```

```python
from sklearn.preprocessing import StandardScaler #mean = 0 and SD = 1 
sc = StandardScaler()
sc.fit(X_train)                                                                                   # fit the data to Standardscaler

X_tests = sc.transform(X_test) 
X_trains = sc.transform(X_train)    # transform the data

print('After standardizing our features, the first 5 rows of our data now look like this:\n')   # print the first 5 value after standardization
print(pd.DataFrame(X_trains, columns=X_train.columns).head()) 
```

```python
print(X_train.shape)
print(Y_train.shape)
print(X_test.shape)
print(Y_train.shape)
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
logreg.fit(X_trains,Y_train)
```

```python
Y_pred_train = logreg.predict(X_trains)  
```

```python
Y_pred_test = logreg.predict(X_tests)
```

```python
Y_pred_test
```

```python
from sklearn.metrics import confusion_matrix

confusion_matrix = pd.DataFrame(confusion_matrix(Y_test, Y_pred_test))

print(confusion_matrix)
```

```python
confusion_matrix.index = ['Actual Lost','Actual Won']
confusion_matrix.columns = ['Predicted Lost','Predicted Won']
print(confusion_matrix)
```

### Decision Tree Classifier

```python
from sklearn import tree
model = tree.DecisionTreeClassifier(random_state = 0)
model.fit(X_trains, Y_train)
```

```python
y_pred_train = model.predict(X_trains)  
```

```python
y_pred_test = model.predict(X_tests)                                                           # make predictions on the testing set
```

```python
from sklearn.metrics import accuracy_score
print('Accuracy score for test data is:', accuracy_score(Y_test,y_pred_test))
```

```python
from sklearn.metrics import confusion_matrix

confusion_matrix = pd.DataFrame(confusion_matrix(Y_test, y_pred_test))

confusion_matrix.index = ['Actual Won','Actual Lost']
confusion_matrix.columns = ['Predicted Won','Predicted Lost']
print(confusion_matrix)
```

### Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(random_state = 0)

model.fit(X_trains, Y_train)
```

```python
model1 = RandomForestClassifier(random_state = 0,
                                n_estimators = 15,
                                min_samples_split = 2,
                                min_samples_leaf = 1)

model1.fit(X_trains, Y_train)
```

```python
Y_pred_train = model.predict(X_trains)
Y_pred_train1 = model1.predict(X_trains)
```

```python
Y_pred_test = model.predict(X_tests)                                                           # make predictions on the testing set
Y_pred_test1 = model1.predict(X_tests)
```

```python
from sklearn.metrics import accuracy_score
print('Accuracy score for test data using the model without parameter specification:', accuracy_score(Y_test,Y_pred_test))
print('Accuracy score for test data using the model with parameter specification:', accuracy_score(Y_test,Y_pred_test1))
```

```python
from sklearn.metrics import confusion_matrix

confusion_matrix = pd.DataFrame(confusion_matrix(Y_test, Y_pred_test))

confusion_matrix.index = ['Actual Won','Actual Lost']
confusion_matrix.columns = ['Predicted Won','Predicted Lost']
print(confusion_matrix)
```
